```python
# necessary prolog
from pyspark import SparkContext
sc = SparkContext()
# neceesary prolog for dataframes
from pyspark import sql
sqlContext = sql.SQLContext(sc)
```

```python
tweets = sqlContext.read.json("1000tweets.json") 

tweets.printSchema()
```

```python
#Krijgt alle tweets met geotags
tweets.filter(tweets["coordinates"].isNotNull()).select(tweets["coordinates"]["coordinates"], tweets["coordinates"]["type"], tweets["user"]["id"], tweets["user"]["name"]).show()
```

```python
#Print de cordinaten en het type cordinaat.
tweets.registerTempTable("tweets");
sqlContext.sql('''
SELECT t.coordinates.coordinates, t.coordinates.type
FROM tweets t
WHERE t.coordinates IS NOT NULL
''').show()
```
